```python
def load_documents(pdf):
    loader=DirectoryLoader(pdf, glob="*.pdf", loader_cls=PyPDFLoader)
    document=loader.load()
    return document
```

```python
text=load_documents("data")
```

```python
def raw_to_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=700, chunk_overlap=50)
    chunks=text_splitter.split_documents(text)
    return chunks
```

```python
chunks=raw_to_chunks(text)
```

```python
def huging_face_embeddings():
    embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    return embeddings
```

```python
embeddings=huging_face_embeddings()
```

```python
lent=embeddings.embed_query("my name is ilias?")
len(lent),lent
```

```python
vec_db = PineconeVectorStore(
    pinecone_api_key=PINECONE_KEY,
    embedding=embeddings,  
    index_name="medicalbot",
)
vec_db.add_texts(texts=[text.page_content for text in chunks])
```
